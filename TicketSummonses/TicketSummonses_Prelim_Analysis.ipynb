```python
import seaborn
import numpy as np
import pandas as pd
import scipy.stats as sc
import matplotlib.pylab as pl
from IPython.display import display
%matplotlib inline
pd.options.display.max_columns = None
```

```python
Tickets = {}
Tickets['y2011'] = pd.read_excel('citysum2011.xlsx', header = 3).dropna(how='any').transpose()
Tickets['y2012'] = pd.read_excel('citysum2012.xlsx', header = 3).dropna(how='any').transpose()
Tickets['y2013'] = pd.read_excel('citysum2013.xlsx', header = 3).dropna(how='any').transpose()
Tickets['y2014'] = pd.read_excel('citysum2014.xlsx', header = 4).dropna(how='any').transpose()
Tickets['y2015'] = pd.read_excel('citysum2015.xlsx', header = 4).dropna(how='any').transpose()
Tickets['y2016'] = pd.read_excel('citysum2016.xlsx', header = 4).dropna(how='any').transpose()
for df in Tickets:
    Tickets[df].columns = Tickets[df].iloc[0]
    Tickets[df] = Tickets[df].drop(['Offense'])
frames = [Tickets['y2011'], Tickets['y2012'], Tickets['y2013'], Tickets['y2014'], Tickets['y2015'], Tickets['y2016']]
df = pd.concat(frames)
df = df.drop(['MTD 2011'])
df = df.drop(['MTD 2012'])
df = df.drop(['MTD 2013'])
df = df.drop(['MTD 2014'])
df = df.drop(['MTD 2015'])
df = df.drop(['MTD 2016'])
```

```python
df['Year'] = pd.Series(np.array([2011, 2012, 2013, 2014, 2015, 2016])).values
df = df.drop(['YTD 2016'])
df['total_sum'] = df[df.columns].sum(axis = 1)
```

```python
# Tickets that should affect collision rates:
dangerticks = ['Disobey_Sign', 'Fail_to_Stop_on_Signal', 'Improper_Turn',
              'Not_Giving_RofW_to_Veh', 'Speeding', 'Fail_to_Signal', 
              'Following_Too_Closely', 'Improper_Passing', 'Not_Giving_RofW_to_Pedes',
              'Unsafe_Lane_Change']
dangerticks
```

```python
df['dangerous_tickets'] = df[dangerticks].sum(axis = 1)
```

```python
df
```

```python
fig1, (ax1) = pl.subplots(nrows = 1, ncols = 1, figsize = (10, 4))
ax1.get_xaxis().get_major_formatter().set_useOffset(False)
width = 0.8
ax1.bar(df.Year[0:3], df.dangerous_tickets[0:3], color = 'k', alpha = 0.4,
         label = 'Risky Tickets Pre-Vision Zero', align = 'center')
ax1.bar(df.Year[3:5], df.dangerous_tickets[3:5], color = 'r', alpha = 0.4,
         label = 'Risky Tickets Under Vision Zero', align = 'center')
ax1.set_ylabel('Potential for Collision Tickets')
ax1.set_xlabel('Year')
ax1.set_title('NYC Moving Violations Ticket Summonses For Collision Relevant Offenses')
ax1.axis([2010.5, 2015.5, 0, 600000])
ax1.legend(loc = 'upper left')
```

```python
prevz = df.dangerous_tickets[0:3].mean()
vz = df.dangerous_tickets[3:5].mean()
```

```python
x = sc.ttest_ind(df.dangerous_tickets[0:3], df.dangerous_tickets[3:5], equal_var = False)
if x[1] < 0.05:
    print 'At p=0.05 we can reject the null hypothesis that there is no difference ' + \
    'or a decrease in the number of dangerous driving tickets issued by the NYPD. ' + \
    'Accordingly, we can accept the alternative hypothesis, where there was an increase ' + \
    'in tickets issued for relevant offenses after Vision Zeros beginning'
else:
    print 'At p=0.05 we cannot reject the null hypothesis: there is no difference ' + \
    'or a decrease in the number of dangerous driving tickets issued by the NYPD.'
```
